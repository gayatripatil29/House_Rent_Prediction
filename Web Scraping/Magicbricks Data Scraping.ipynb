```python
from bs4 import BeautifulSoup as soup
from selenium import webdriver
import requests
import pandas as pd
```

```python
# For URL's
df=pd.read_excel('Links.xlsx',header=None)
url_list=df
```

```python
name=[]
bedroom=[]
bathroom=[]
facing=[]
super_area=[]
carpet_area=[]
furnished_status=[]
car_parking=[]
water_availability=[]
floor=[]
rental_value=[]
security_deposit=[]
monthly_charges=[]
brokerage=[]
status=[]
locality=[]
locality_rating=[]
age_of_construction=[]
over=[]
tenant=[]
owner=[]
links=[]
```

```python
import time
counter = 0
num=0
for i in range(len(url_list[0][5])):
    link=url_list[0][i]
    num=num+1
    print(num,link)
    # Chrome Driver
    driver=webdriver.Chrome()
    driver.get(link)
    source=driver.page_source
    page_soup=soup(source,'html.parser')
    
    

        # Name
    try:
        nam=(page_soup.find('div',{"id":"propUnitDesc","style":"display:none;"}).text).replace('\n','').strip()
        name.append(nam) # List of Names
        avail=True
    except AttributeError:
        avail=False
        print('Link problem')
        pass

       
    if avail==True:
        
        links.append(link)
        #Bedroom

        bed=page_soup.find('div',{'class':"seeBedRoomDimen"})
        if not bed:
            bedroom.append('NA')
        else:
            bedroom.append(bed.text.replace('See Dimension','').replace('Bedroom','').replace('\n','').strip()[0])

        #Bathroom
        aj=(page_soup.findAll('div',{'class':"p_infoColumn"}))
        for j in aj:
            aq=j.find('div',{'class':"p_title"}).text.strip()   
            if aq=='Bathroom' or aq=='Bathrooms':
                as1 = j.find('div',{'class':"p_value"})
                if not as1:
                    bathroom.append('NA')
                else:
                    bathroom.append(as1.text)

        #Super_area
        sup_area=page_soup.find('div',{"id":"coveredArea","style":"display: none;"})
        if not sup_area:
            super_area.append('NA')
        else:
            super_area.append(sup_area.text.replace('\n','').strip())

        #Carpet area
        car_area=page_soup.find('div',{"id":"carpetArea","style":"display: none;"})
        if not car_area:
            carpet_area.append('NA')
        else:
            carpet_area.append(car_area.text.replace('\n','').strip())

        #Furnished status
        fur=(page_soup.findAll('div',{'class':"p_infoColumn"}))
        for j in fur:
            fur_stt=j.find('div',{'class':"p_title"}).text.strip()   
            if fur_stt=='Furnished status':
                furn_status = j.find('div',{'class':"p_value"})
                if not furn_status:
                    furnished_status.append('NA')
                else:
                    furnished_status.append(furn_status.text)

        #Car parking
        car_status=False
        car=(page_soup.findAll('div',{'class':"p_infoColumn"}))
        for j in car:
            car_p=j.find('div',{'class':"p_title"}).text.strip()   
            if car_p=='Car parking':
                car_pkg = j.find('div',{'class':"p_value"})
                car_status=True
                             
        if car_status==True:
                if not car_pkg:
                    car_parking.append('NA')
                else:
                    car_parking.append(car_pkg.text) 
        else:
            car_parking.append('NA')

        #Water availability
        water_status=False
        wat=(page_soup.findAll('div',{'class':"p_infoColumn"}))
        for j in wat:
            wat_st=j.find('div',{'class':"p_title"}).text.strip()   
            if wat_st=='Water Availability':
                water_stt = j.find('div',{'class':"p_value"})
                water_status=True
                
        if water_status==True:
            if not water_stt:
                water_availability.append('NA')
            else:
                water_availability.append(water_stt.text)
        else:
            water_availability.append('NA')

        #Floor
        floor_status=False
        fl=(page_soup.findAll('div',{'class':"p_infoColumn"}))
        for j in fl:
            f=j.find('div',{'class':"p_title"}).text.strip()   
            if f=='Floor':
                floor_num = j.find('div',{'class':"p_value"})
                floor_status=True
                
        if floor_status==True:
            if not floor_num:
                floor.append('NA')
            else:
                 floor.append(floor_num.text)
                    
        else:
            floor.append('NA')

        # Rent,security deposit Monthly charges, Brokerage
        rent_val=False
        mchg_val=False
        secdep_val=False
        brok_val=False

        chgs=page_soup.find('div',{'class':'srpdetailWrap','id':'srpdetailWrap'})
        a=chgs.findAll('div',{'class':'td'})

        for i in range(len(a)):
            if a[i].text=='Rent' or 'Monthly Rent' in a[i].text:
                rental_value.append(a[i+1].text.strip())
                rent_val=True   
            if a[i].text=='Monthly Maintenance':
                monthly_charges.append(a[i+1].text.strip())
                mchg_val=True      
            if a[i].text=='Security Deposit':
                security_deposit.append(a[i+1].text.strip())
                secdep_val=True
            if 'Brokerage' in a[i].text:
                brokerage.append(a[i+1].text.strip())
                brok_val=True

         
        if rent_val==False:
            try:
                rr=page_soup.findAll('div',{'class':"fltLeft"})
                for i in rr:
                    a=i.findAll('span',{'class':'semiBold'})
                    for j in range(len(a)):
                        if j==0:
                            rental_value.append(a[j].text.strip().replace('₹','').replace('\n',''))
            except AttributeError:
                rental_value.append('NA')
        if mchg_val==False:
            rr=page_soup.findAll('div',{'class':"fltLeft"})
            for i in rr:
                a=i.findAll('span',{'class':'semiBold'})
                for j in range(len(a)):
                    if j==1:
                        monthly_charges.append(a[j].text.strip().replace('₹','').replace('\n',''))
                        mchg_val=True
            if mchg_val==False:
                monthly_charges.append('NA')
        if secdep_val==False:
            try:
                sec=page_soup.findAll('div',{'id':'bookingAmount','class':'p_infoRow p_infoRowSecDep'})
                sec_deposit=sec[0].find('div',{'class':'breakupdivider'}).text.strip().replace('₹','').replace('\n','')
                security_deposit.append(sec_deposit)
            except AttributeError:    
                security_deposit.append('NA')
        if brok_val==False:
            brokerage.append('NA')

        #Status
        stat_status=False
        stat=page_soup.find('div',{'class':'descriptionCont'})
        a1=stat.findAll('div',{'class':'p_title'})
        for i in range(len(a1)):
            if a1[i].text=='Status':
                st=stat.findAll('div',{'class':'p_value'})[i]
                stat_status=True
                
        if stat_status==True:
            if not st:
                status.append('NA')
            else:
                status.append(st.text)
                
        else:
            status.append('NA')

        #Locality
        locality.append(url_list[1][i])


        #age_of_construction=[]
        const_status=False
        age=page_soup.find('div',{'class':'descriptionCont'})
        a=age.findAll('div',{'class':'p_title'})
        for i in range(len(a)):
            if a[i].text=='Age of Construction':
                p_v=age.findAll('div',{'class':'p_value'})[i]
                const_status=True
        if const_status==True:
            if not p_v:
                age_of_construction.append('NA')
            else:
                age_of_construction.append(p_v.text)
        else:
            age_of_construction.append('NA')

        #overlooking
        o_status=False
        o=page_soup.find('div',{'class':'descriptionCont'})
        a1=o.findAll('div',{'class':'p_title'})
        for i in range(len(a1)):
            if a1[i].text=='Overlooking':
                ov=o.findAll('div',{'class':'p_value'})[i].text
                o_status=True
        if o_status==True:
            if not ov:
                over.append('NA')
            else:
                over.append(ov)
        else:
            over.append('NA')

        #facing
        facing_status=False
        fac=(page_soup.findAll('div',{'class':"p_infoColumn"}))
        for j in fac:
            face=j.find('div',{'class':"p_title"}).text.strip()   
            if face=='Facing':
                direction = j.find('div',{'class':"p_value"}).text
                facing_status=True
                
        if facing_status==True:
            if not direction:
                facing.append('NA')
            else:
                facing.append(direction)
        else:
            facing.append('NA')

        #tenant
        tenant_status=False
        ten=(page_soup.findAll('div',{'class':"p_infoColumn"}))
        for j in ten:
            t=j.find('div',{'class':"p_title"}).text.strip()   
            if t=='Tenants Preferred':
                tent = j.find('div',{'class':"p_value"}).text
                tenant_status=True
        if tenant_status==True:
            if not tent:
                tenant.append('NA')            
            else:
                tenant.append(tent)
        else:
            tenant.append('NA')

        #owner
        
        own=page_soup.find('div',{'class':'nameTitle'})
        if not own:
            owner.append('Agent')
        elif own.text=='Owner':
            owner.append(own.text)
        else:
            owner.append('NA')

            
            
    else:
        counter=+1
        pass 
        

        

    
# For building Dataframe

# All Independent Variable Dataframes
a=pd.DataFrame(name)
b=pd.DataFrame(bedroom)
c=pd.DataFrame(bathroom)
d=pd.DataFrame(super_area)
e=pd.DataFrame(carpet_area)
f=pd.DataFrame(furnished_status)
g=pd.DataFrame(car_parking)
h=pd.DataFrame(water_availability)
i=pd.DataFrame(floor)
j=pd.DataFrame(rental_value)
k=pd.DataFrame(security_deposit)
l=pd.DataFrame(status)
m=pd.DataFrame(locality)
n=pd.DataFrame(age_of_construction)
o=pd.DataFrame(over)
p=pd.DataFrame(facing)
q=pd.DataFrame(monthly_charges)
r=pd.DataFrame(brokerage)
s=pd.DataFrame(tenant)
t=pd.DataFrame(owner)
u=pd.DataFrame(links)
main_df=pd.concat([a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u],axis=1)
my_col=['Name','Bedroom','Bathroom','Super_area','Carpet_area','Furnished_status','Car_parking','Water_availability','Floor','Rent','Security_deposit','Availability_status','Locality','Age_of_construction','Overlooking','Facing','Monthly_charges','Brokerage','Tenant','Owner','Links']
main_df.columns=my_col
main_df.to_csv("regression_final.csv")
```
